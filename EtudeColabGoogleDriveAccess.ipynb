# ColabからGoogleDriveにアクセス

```python
from google.colab import drive
drive.mount('/content/drive')
```

## Colab Notebook のファイルリストを取得

```python
import glob

files = glob.glob('drive/My Drive/Colab Notebooks/*.ipynb', recursive=True)
for file in files:
    print(file.split("/")[-1].strip(".ipynb"))
```



---



## ファイルリストの取得（全ファイル）

```python
import glob

files = glob.glob('drive/My Drive/**/*', recursive=True)
for file in files:
    print(file.split("/")[-1].strip("."))
```

## ファイルリストの取得（pdfのみ）

```python
import glob

files = glob.glob('drive/My Drive/**/*.pdf', recursive=True)
for file in files:
    print(file.split("/")[-1].strip(".pdf"))
```

## ファイルリストの取得（mdのみ）

```python
import glob

files = glob.glob('drive/My Drive/**/*.md', recursive=True)
for file in files:
    print(file.split("/")[-1].strip(".md"))
```
